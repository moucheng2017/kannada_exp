```python
# load the data 
train = pd.read_csv('/home/xmc/projects_data/Kannada-MNIST/train.csv')
print(train.shape)
train.head()
```

```python
# load the data 
test = pd.read_csv('/home/xmc/projects_data/Kannada-MNIST/test.csv')
print(test.shape)
test.head()
```

```python
# split the columns as 1st column is the label and the others are digits
x_train_df = train.iloc[:,1:]
y_train_df = train.iloc[:,0]

print(x_train_df.shape, y_train_df.shape)
```

```python
# split the columns as 1st column is the label and the others are digits
x_test_df = test.iloc[:,1:]
id_test_df = test.iloc[:,0]

print(x_test_df.shape, id_test_df.shape)
```

```python
# check the means of all of the digits values
# mean_val = []
# std_val = []
```

```python
# normalisation:
x_train = x_train_df.values/255.
y_train = y_train_df.values

x_test = x_test_df.values/255
```

```python
# reshape the images
x_train = np.reshape(x_train, (60000,1,28,28))
x_test = np.reshape(x_test, (5000, 1,28,28))

x_train.shape, x_test.shape
```

```python
def display(rows, columns, images, values=[], predictions=[]):
    fig = plt.figure(figsize=(9, 11))

    ax = []

    for i in range( columns*rows ):
        img = images[i]
        ax.append(fig.add_subplot(rows, columns, i+1))
        
        title = ""
        
        if(len(values) == 0):
            title = "Pred:" + str(predictions[i])
        elif(len(predictions) == 0):
            title = "Value:" + str(values[i])
        elif(len(values) != 0 and len(predictions) != 0):
            title = "Value:" + str(values[i]) + "\nPred:" + str(predictions[i])
        
        ax[-1].set_title(title)  # set title
        plt.imshow(img)

    plt.show()
    
idx = np.random.randint(1, 1000, size=9)

images = x_train[idx,:]
images = images[:,0]

values = y_train[idx]

display(rows=3, columns=3, images=images, values=values, predictions=[])
```
